```python
class lentaRu_parser:
    def __init__(self):
        pass


    def _get_url(self, param_dict: dict) -> str:
        url = (
          f"https://lenta.ru/search/v2/process?"
          f"from={param_dict['from']}&"
          f"size={param_dict['size']}&"
          f"sort={param_dict['sort']}&"
          f"title_only={param_dict['title_only']}&"
          f"domain={param_dict['domain']}&"
          "modified%2Cformat=yyyy-MM-dd&"
          f"modified%2Cfrom={param_dict['dateFrom']}&"
          f"modified%2Cto={param_dict['dateTo']}&"
          f"query={param_dict['query']}")
        if param_dict['type'] != 0:
          url += f"&type={param_dict['type']}"
        if param_dict['bloc'] != 0:
          url += f"&bloc={param_dict['bloc']}"
        return url



    def _get_search_table(self, param_dict: dict) -> pd.DataFrame:
        """
        Возвращает pd.DataFrame со списком статей
        """
        url = self._get_url(param_dict)
        r = rq.get(url)
        search_table = pd.DataFrame(r.json()['matches'])

        return search_table


    def get_articles(self,
                     param_dict,
                     time_step = 50,
                     save_every = 5,
                     save_excel = True) -> pd.DataFrame:
        """
        Функция для скачивания статей интервалами через каждые time_step дней
        Делает сохранение таблицы через каждые save_every * time_step дней

        param_dict: dict
        ### Параметры запроса
        ###### project - раздел поиска, например, rbcnews
        ###### category - категория поиска, например, TopRbcRu_economics
        ###### dateFrom - с даты
        ###### dateTo - по дату
        ###### offset - смещение поисковой выдачи
        ###### limit - лимит статей, максимум 100
        ###### query - поисковой запрос (ключевое слово), например, РБК

        """
        param_copy = param_dict.copy()
        time_step = timedelta(days=time_step)
        dateFrom = datetime.strptime(param_copy['dateFrom'], '%Y-%m-%d')
        dateTo = datetime.strptime(param_copy['dateTo'], '%Y-%m-%d')
        if dateFrom > dateTo:
            raise ValueError('dateFrom should be less than dateTo')

        out = pd.DataFrame()
        save_counter = 0

        while dateFrom <= dateTo:
            param_copy['dateTo'] = (dateFrom + time_step).strftime('%Y-%m-%d')
            if dateFrom + time_step > dateTo:
                param_copy['dateTo'] = dateTo.strftime('%Y-%m-%d')
            print('Parsing articles from '\
                  + param_copy['dateFrom'] +  ' to ' + param_copy['dateTo'])
            out = pd.concat([out, self._get_search_table(param_copy)], axis=0, ignore_index=True)
            dateFrom += time_step + timedelta(days=1)
            param_copy['dateFrom'] = dateFrom.strftime('%Y-%m-%d')
            save_counter += 1
            if save_counter == save_every:
                display.clear_output(wait=True)
                out.to_excel("/tmp/checkpoint_table.xlsx")
                print('Checkpoint saved!')
                save_counter = 0

        if save_excel:
            out.to_excel("lenta_{}_{}.xlsx".format(
                param_dict['dateFrom'],
                param_dict['dateTo']))
        print('Finish')

        return out
```

### Word2Vec

Учебное пособие по Gensim Word2Vec: https://www.kaggle.com/code/pierremegret/gensim-word2vec-tutorial

**Параметры модели:**
* `min_count` = int - Игнорирует все слова с суммарной абсолютной частотой ниже этого значения - (2, 100)
* `window` = int - Максимальное расстояние между текущим и прогнозируемым словом в предложении. Например, слова окна слева и слова окна слева от нашей цели - (2, 10)
* `size` = int - Размерность векторов признаков. - (50, 300)
* `sample` = float - Пороговое значение для настройки того, какие слова с более высокой частотой будут случайным образом отбираться с пониженной частотой. Очень важно. - (0, 1e-5)
* `alpha` = float - Начальная скорость запоминания - (0,01, 0,05)
* `min_alpha` = float - скорость обучения будет линейно снижаться до minus_alpha по мере прохождения обучения. Чтобы задать ее: `alpha - (min_alpha * epochs) ~ 0,00`
* `negative` = int - Если > 0, будет использоваться отрицательная выборка, значение int для negative указывает, сколько "шумовых слов" должно быть выведено. Если установлено значение 0, отрицательная выборка не используется. - (5, 20)
* `workers` = int - Используйте это множество рабочих потоков для обучения модели (=более быстрое обучение на многоядерных машинах).


**Параметры обучения:**

* `total_examples` = int - Количество предложений;
* `epochs` = int - Количество итераций (эпох) по корпусу - [10, 20, 30]

```python
sent = [row.split() for row in X_train['text']]

HIDDEN = 100

w2v = Word2Vec(min_count=10,
                     window=2,
                     vector_size=HIDDEN,
                     sample=6e-5,
                     alpha=0.03,
                     min_alpha=0.0007,
                     negative=20,
                     workers=2)
w2v.build_vocab(sent, progress_per=10000)
w2v.train(sent, total_examples=w2v.corpus_count, epochs=30, report_delay=1)
```

```python
# Cчитаем средний вектор текста
def get_mean_w2v_vector(sentence):
    Sum = 0
    Count = 0

    try:
      words = sentence.split()
    except TypeError:
      words = []

    for w in words:
        if w in w2v.wv:
            Sum += w2v.wv[w]
            # Sum += glove_vectors[w]
            Count += 1

    if Count == 0:
        return 0

    return Sum / Count
```

```python
NewCols = ['col'+str(i) for i in range(HIDDEN)]

X_train['vectors'] = X_train['text'].map(get_mean_w2v_vector)
X_test['vectors'] = X_test['text'].map(get_mean_w2v_vector)

X_train.head()
```

```python
# Удалим строки, для которых векторы не посчитались (все слова оказались неизвестными)
IdxTrain = []

for ix, row in X_train.iterrows():
    if not isinstance(row['vectors'],np.ndarray):
        IdxTrain.append(ix)

IdxTest = []

for ix, row in X_test.iterrows():
    if not isinstance(row['vectors'],np.ndarray):
        IdxTest.append(ix)
```

```python
X_train_w2v = X_train.drop(index=IdxTrain)
X_test_w2v = X_test.drop(index=IdxTest)

y_train_w2v = y_train.drop(index=IdxTrain)
y_test_w2v = y_test.drop(index=IdxTest)
```

```python
X_train_w2v[NewCols] = pd.DataFrame(X_train['vectors'].tolist(), index=X_train_w2v.index)
X_test_w2v[NewCols] = pd.DataFrame(X_test['vectors'].tolist(), index=X_test_w2v.index)
X_train_w2v.drop(columns = ['text','vectors'], inplace=True)
X_test_w2v.drop(columns = ['text','vectors'], inplace=True)

X_train_w2v.head()
```

#### Логистическая регрессия

```python
logreg_w2v = LogisticRegression(max_iter=300, class_weight = 'balanced', random_state=42)
logreg_w2v.fit(X_train_w2v, y_train_w2v)
y_pred = logreg_w2v.predict(X_test_w2v)

print(classification_report(y_test_w2v, y_pred))
```

### Предобученный BERT

```python
# For DistilBERT:
model_class, tokenizer_class, pretrained_weights = (ppb.DistilBertModel, ppb.DistilBertTokenizer, 'distilbert-base-uncased')

## Want BERT instead of distilBERT? Uncomment the following line:
#model_class, tokenizer_class, pretrained_weights = (ppb.BertModel, ppb.BertTokenizer, 'bert-base-uncased')

# Load pretrained model/tokenizer
tokenizer = tokenizer_class.from_pretrained(pretrained_weights)
model = model_class.from_pretrained(pretrained_weights)
```

```python
# токенизация на слова и подслова в BERT-формате
tokenized = X['text'].apply((lambda x: tokenizer.encode(x, add_special_tokens=True)))
```

```python
# Padding. Мы будем подавать данные в BERT батчами (группами), поэтому все тексты должны быть одинаковой длины - дополним короткие предложения нулями

max_len = 512
padded = np.array([i[:max_len] + [0]*(max_len-len(i)) for i in tokenized.values])
```

```python
# Masking. Нужно показать BERT-у, что нули - это пустые символы
attention_mask = np.where(padded != 0, 1, 0)
attention_mask.shape
```

```python
# Необходимые нам векторы находятся в переменной last_hidden_states
input_ids = torch.tensor(padded)
attention_mask = torch.tensor(attention_mask)

with torch.no_grad():
    last_hidden_states = model(input_ids, attention_mask=attention_mask)
```

```python
X_bert = last_hidden_states[0][:,0,:].numpy() # векторы текстов длины 768
```

```python
X_train_bert, X_test_bert, y_train_bert, y_test_bert = train_test_split(X_bert, y, test_size = 0.25, random_state=12345)
```

#### Логистическая регрессия

```python
logreg_bert = LogisticRegression(max_iter=300, class_weight = 'balanced', random_state=42)
logreg_bert.fit(X_train_bert, y_train_bert)
y_pred = logreg_bert.predict(X_test_bert)

print(classification_report(y_test_bert, y_pred))
```
